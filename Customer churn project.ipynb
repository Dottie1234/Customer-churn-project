```python
import pandas as pd

customer_file = pd.read_csv('Customer-Churn-Records.csv')
customer_file
```

##### 1. find mean credit score of the customers

```python
customer_file['CreditScore'].mean()
```

##### 2a. Provide the surname, geography and gender of the customer with the highest credit score

```python
max_credit = customer_file['CreditScore'].max()
high_credit_score = customer_file[customer_file['CreditScore'] == max_credit].loc[:, ['CreditScore', 'Surname', 'Geography', 'Gender', 'Age']]
high_credit_score
```

##### 2b. output information of the oldest person with the highest credit score.

```python
oldest_high_credit = high_credit_score['Age'].max()
high_credit_score[high_credit_score['Age'] == oldest_high_credit]
```

##### 2c. output the information of the youngest person with the highest credit score

```python
youngest_high_credit = high_credit_score['Age'].min()
high_credit_score[high_credit_score['Age'] == youngest_high_credit]
```

##### 3a. Provide the surname, geography and gender of the customer with the lowest credit score.

```python
low_credit = customer_file['CreditScore'].min()
low_credit_score = customer_file[customer_file['CreditScore'] == low_credit].loc[:, ['CreditScore', 'Surname', 'Geography', 'Gender', 'Age']]
low_credit_score
```

##### 3b. output the information of the oldest person with the lowest credit score

```python
oldest_low_credit = low_credit_score['Age'].max()
low_credit_score[low_credit_score['Age'] == oldest_low_credit]
```

##### 3c. output the information of the youngest person with the lowest credit score

```python
youngest_low_credit = low_credit_score['Age'].min()
low_credit_score[low_credit_score['Age'] == youngest_low_credit]
```

##### 4. Create a new column called 'age bins' and group assign each row to a particular age bin using the age column.

```python
customer_file['Age bins'] = customer_file['Age'].apply(lambda x: 'Teenager' if x <= 18 else 'Adult' if x <= 30 else 'Middle aged' if x <= 50 else ' Old')
customer_file
```

##### 5. Using the age_bin column find the most common age bins of churned customers.

```python
customer_file['Age bins'].value_counts()
```

##### 6. Using the 'Estimated Salary of customer' and Active member of the bank column, find the mean, median and mode of the estimated salaries of customers who are active

```python
#active customers have the value of 1
active_customers = customer_file[customer_file['IsActiveMember'] == 1]
print(f"The mean of active customer's estimated salary is {active_customers['EstimatedSalary'].mean()}")
print(f"The median of active customer's estimated salary is {active_customers['EstimatedSalary'].median()}")
print(f"The mode of the active customer's estimated salary is {active_customers['EstimatedSalary'].mode()[0]}")
```

##### 7. using the estimated salary of customer' and Active member of the bank column, find the mean, median and mode of the estimated salaries of customers who are not active

```python
#non active customers have the value of 0
non_active_customers = customer_file[customer_file['IsActiveMember'] == 0]
print(f"The mean of non active customer's estimated salary is {non_active_customers['EstimatedSalary'].mean()}")
print(f"The median of non active customer's estimated salary is {non_active_customers['EstimatedSalary'].median()}")
print(f"The mode of the non active customer's estimated salary is {non_active_customers['EstimatedSalary'].mode()[0]}")
```

##### 8. what is the percentage of customers who has a complaint?

```python
percentage = len(customer_file.query('Complain == 1'))/len(customer_file)*100
print(percentage)
```

##### 9. What is the percentage of customers who didnt have a complaint?

```python
percentage = len(customer_file.query('Complain == 0'))/ len(customer_file)*100
print(percentage)
```
